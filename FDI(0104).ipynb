### <b id ='b2'> Introduction
After the reform and opening-up policy was carried out in 1979, the national economic system in China transmitted from a planned economy to market-oriented economy. Before the advent of reform in 1979, there was almost no FDI. However, by the end of 2008, the accumulated FDI inflows was up to 1099.8 billion US dollars, and the annual average growth rate was 20.3%. China has already surpassed USA and become the largest recipient of FDI. However, most of the attracted FDI has flown into East China.By the end of 2008, the accumulated FDI inflow(1979 - 2008) in China was amount to 1099.8 billion dollars, 81.09% of which were concentrated in East China. East China is the coastal region. It is always the main gathering place of FDI, thanks to the first implementation of opening-up policy, better social and economic foundation, more stable market economy, and excellent geographical location. Furthermore, in the year of 2008, the economy size of eastern part was the largest, accounting for 58% in the gross domestic product (GDP).<br/>

In this study, we will discover the key influencing factors of FDI inflows in East China, which is the most developed region in China.<br/>

In 1986, the seventh Five-Year Plan from the Sixth National People's Congress divided 31 provinces in Mainland China into 3 parts, eastern part, middle part and western part. Thereinto, the eastern part includes 11 provinces --- Beijing, Tianjin, Hebei, Liaoning, Shanghai, Jiangsu, Fujian, Shandong, Guangdong, Hainan, and Zhejiang.<br/>

### <b id ='b3'> Empirical Model
On the basis of existing research findings and the actual situation of FDI in China, 5 influencing factors are selected in this study. We will study these influencing factors of attracted FDI into East China.<br/>
#### 1. Data Sources

All the data used here is from"Chinese Statistical Yearbook from 1949 - 2008","Chinese International Trade Statistical Yearbook". The total volumn of foreign trade are converted to RMB by the annual exchange rate of RMB (1987 - 2008) against US dollar. Besides, we adopt the amount of foreign capital actually utilized in this study.<br/>

In the begining years of reform and opening up, total amount of FDI inflows was really small, in light of this, data during 1987 - 2008 will be chosen. <br/>
#### 2. Explanatory Variables

In this study, after weighing data availability and reliability, the following explanatory variables will be chosen:<br/>
1ï¼‰ **Expenditure**, the government expenditure<br/>
By the end of 2008, the government procurement has a huge impact on multinational enterprises(MNEs). On the one hand, some governments normally prefer pruchasing from domestic manufacturers.One likely reason is to protect domestic infant industries, but creating unfair competition environment for MNEs. One the other hand, the government spending includes various aspects, such as, facilities, medical services, social security, education, national defence, and environmental protection, etc.We have no reason to believe that government spending will hinder economic growth and FDI inflows. Especially for some proverty-stricken areas, the government spending is a must. <br/>
So, we do not make any expectations on this variable. <br/><br/>
2ï¼‰ **Thirdrate**, the ratio of tertiary industry output to the output of all industries<br/>
This variable reflects the level of service industry, as well as the level of modernization. The research literature did not cover the impact of tertiary industry on FDI inflows very well, most of the existing literature discusses from the other direction --- FDI inflows will improve the tertiary industry's proportion. <br/>
Here, we expect that tertiary industry is positively correlated with FDI. <br/><br/>
3ï¼‰ **SErate**, the ratio of the output of state-owned enterprises(SOEs) to gross output value of all enterprises, measuring marketization level in one region<br/>
Mature market mechanism can help MNEs reduce investment risks, information costs, and costs caused by government intervention. But high proportion of SOEs in one region will distort market mechanism, adverse to fair competition(Huang,2006)<sup>1</sup>.<br/>
Due to the diffuculty of measuring the marketization level in one area, most of the scholars use SErate, which can roughly show the development of market economy in one country or region. <br/>
Here, we expect SErate is negatively correlated with FDI inflows. <br/><br/>
4ï¼‰ **FA**, referring to the total investment in fixed assets. <br/>
Here it is used as a proxy for infrastructure level in one region. Impeccable infrastructure is necessary for one country or region to attract FDI(Dunning, 1977)<sup>2</sup>, as infrastructure in one region reflects the convenience degree of operating business, obtaining information and raw materials, transportation of finished goods, ets.For going into production smoothly, foreign investors need those infrastructure which is indispensable for industrial activities, such transportation, telecommunication, hydroelectricity etc. But they cannot construct these facilities themselves, local government need to prepare well for them. <br/>
The effect from this factor has already been confirmed in many studies, here we expected there is a strongly positive correlation between FA and FDI inflows. <br/><br/>
5ï¼‰ **Wage**, measurement of labor cost.<br/>
It is widely believed that lower labor cost is one of the advantage of for developing countries to attract FDI, and there should be a negative relationship between each other(Ymawaki,1991)<sup>3</sup>.However, many existing research is not consistent with this belief. And some show that wage and FDI are positively correlated(Coughlin,1991)<sup>4</sup>. An explanation for this is that labor cost is an indirect measurement of labor force quality. Higher labor cost is an implication of better labor quality, which is favored by MNEs.In some other studies, scholars find that wage level will not affect the appeal of this region to FDI(Broadman et al, 1997)<sup>5</sup>.<br/>
So here, we do not make any expectations on this variable. <br/><br/>
#### 3. Model

The model we build in this study is as follows:<br/>
\begin{equation*}
Log(FDI)=\beta_0 + \beta_1 Log(Expenditure_t) + \beta_2 Log(Thirdrate_t) + \beta_3 Log(SErate_t) + \beta_4 Log(FA_t) + \beta_5 log(Wage_t) + \epsilon_t
\end{equation*}

Where,<br/>
$Expenditure_t$: government expenditure<br/>
$Thirdrate_t$: output of tertiary industry / gross output of all industries<br/>
$SErate_t$: output of state owned enterprises / total output of all enterprises<br/>
$FA_t$: infrastructure level<br/>
$Wage_t$: average wage level<br/>
$epsilon_t$: error term

### <b id ='b4'> Results
The following table is the final result we get. <br/>
**1ï¼‰ Government Expenditure**<br/>
The coefficient is negatively correlated with FDI inflows, and statistically significant at a 1% level,which is perfectly within our expectation. <br/>
Le(2005)<sup>6</sup> mentions that one of the main financial sources for government expenditure is tax, so a potential reason is that those MNEs concern that too much government spending will bring heavier tax to them. Besides, Larrain(2004)<sup>7</sup> addresses that the increasing governments'involvement in public sector will cause higher levels of corruption.<br/>
**2ï¼‰ Tertiary Industry** <br/>
The result is also within our expectation. The coefficient on Thirdrate is positive,and statistically significant at a 1% level, fully illustrating that the role of tertiary industry in developed coastal areas is crucial. <br/>
**3ï¼‰ Marketization Level** <br/>
The coefficient on this explanationary variable is negative, and statistically significant at a 5% level, which is also within our expectation. <br/>
After 1992, more and more MNEs tended to establish wholly foreign-owned enterprises, focusing on occupying local markets, which requires a sound market system. The regions with large proportion SOEs are not attractive to foreign investors anymore, especially in the coastal developed region.<br/>
**4ï¼‰ Social Investment in Fixed Assets** <br/>
Considering the availability of data,social investment in fixed assets will be used as a proxy for the level of infrastructure.<br/>
Empirical result indicates that the improvement of infrastructure can contribute to attracting more FDI, which is also perfectly in line with our expectation.<br/>
**5ï¼‰ Wage** <br/>
Because of the reasons which we mention above, we do not make any expectations on wage.<br/>
The coefficient on wage is positively correlated with FDI inflows, and statistically significant at a 1% levle.The reason we infer is that employers are more willing to pay more to get high-quality labor.

### <b id ='b5'> Conclusion
We carry out regression analysis in this study, through establishment of FDI model with 5 selected explanatory variables.<br/>
The main purpose of this study is to analyze the causes of FDI inflows in East China. The empirical analysis indicates that _infrastructure_,_government expenditure_,third industry_,the degree of marketization_, and _wage_ are all important influencing factors.<br/><br/>
Thus, some policy recommendations are provided as follows:<br/>
1. Tertiary industries has played an important role in the new era, so the governments of the eastern part should focus more on increasing expenses on R&D to attract more foreign investment on tertiary industries, such as, financial industry, IT and media industry. <br/>

2. The local governments in East China should control the government expenditure better and try to make it more transparent.<br/>
